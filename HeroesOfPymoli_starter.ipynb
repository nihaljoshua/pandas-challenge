```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data_df = pd.DataFrame(purchase_data)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
player_count = purchase_data_df["SN"].value_counts()
player_count

summary_df = pd.DataFrame({"Total number of players": [len(player_count)]})
summary_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_item_count = len(purchase_data_df["Item Name"].unique())

average_price = purchase_data_df["Price"].mean()

number_purchases = len(purchase_data)

total_revenue = purchase_data_df["Price"].sum()

summary_df = pd.DataFrame({"Number of Unique Items": [unique_item_count],
                              "Average Price": [average_price],
                              "Number of Purchases": [number_purchases],
                              "Total Revenue": [total_revenue]})
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
total_player = len(purchase_data_df["SN"].value_counts())

male_player = (purchase_data_df.loc[purchase_data_df["Gender"] == "Male"])

Female_player = (purchase_data_df.loc[purchase_data_df["Gender"] == "Female"])
 
other_player = (purchase_data_df.loc[purchase_data_df["Gender"] == "Other / Non-Disclosed"])

male = len(male_player["SN"].value_counts())

female = len(female_player["SN"].value_counts())

other = len(other_player["SN"].value_counts())

male_player_percentage = (male/total_player)*100

female_player_percentage = (female/total_player)*100

other_player_percentage = (other/total_player)*100

summary_df = pd.DataFrame({"Number of Male Players": [male, male_player_percentage],
                              "Number of Female Players": [female, female_player_percentage],
                              "Number of Other Players": [other, other_player_percentage]})
summary_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count_male = male_player["Item Name"].count()
purchase_price_male = male_player["Price"].mean()
purchase_average_male = male_player["Price"].sum()
average_purchase_male = (purchase_average/male)

purchase_count_female = female_player["Item Name"].count()
purchase_price_female = female_player["Price"].mean()
purchase_average_female = female_player["Price"].sum()
average_purchase_female = (purchase_average/female)

purchase_count_other = other_player["Item Name"].count()
purchase_price_other = other_player["Price"].mean()
purchase_average_other = other_player["Price"].sum()
average_purchase_other = (purchase_average/other)


summary_purchase_df = pd.DataFrame({"Gender": ["Male", "Female", "Other / Non-Disclosed"],
                                        "Purchase count": [purchase_count_male, purchase_count_female, purchase_count_other],
                                        "Average Purchase Price": [purchase_price_male, purchase_price_female, purchase_price_other],
                                        "Total purchase value": [purchase_average_male, purchase_average_female, purchase_average_other],
                                        "Avg. Purchase per Person": [average_purchase_male, average_purchase_female, average_purchase_other]})

summary_purchase_df.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
print(purchase_data_df["Age"].max())
print(purchase_data_df["Age"].min())
```

```python
bins = [0, 10, 15, 20, 25, 30, 35, 40, 9999]
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", 
                "30-34", "35-39", "40+"]

purchase_data_by_age = purchase_data_df
purchase_data_by_age["Age Group"] = pd.cut(purchase_data_by_age["Age"], bins, labels=group_labels)
purchase_data_by_age

purchase_data_by_age = purchase_data_by_age.groupby("Age Group")
purchase_data_by_age.count()

summary_data_by_age = pd.DataFrame(purchase_data_by_age.count())
summary_data_by_age

summary_data_by_age["Purchase ID"] = (summary_data_by_age["Purchase ID"]/total_player)*100
summary_data_by_age

summary_data_by_age["Purchase ID"] = summary_data_by_age["Purchase ID"].map("{:,.2f}%".format)
summary_data_by_age

reduce_summary = summary_data_by_age[["Purchase ID", "SN"]]
reduce_summary

rename_summary = reduce_summary.rename(columns={"Purchase ID": "Percentage of Players", "SN": "Total Count"})
rename_summary
```
